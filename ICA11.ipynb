```r
library(ggplot2)
library(dplyr)
library(readr)
file_path <- "/content/drive/MyDrive/Colab Notebooks/DATA/flights.csv"
flights <- read_csv(file_path, show_col_types = FALSE)
flights$month <- factor(flights$month, levels = month.abb, ordered = TRUE)
monthly_flights <- flights %>%
  group_by(year, month) %>%
  summarise(total_passengers = sum(passengers), .groups = 'drop')
ggplot(monthly_flights, aes(x = month, y = total_passengers, color = as.factor(year), group = year)) +
  geom_line(linewidth = 1) +
  geom_point() +
  labs(
    title = "Monthly Flight Trends Over the Years",
    x = "Month",
    y = "Total Passengers",
    color = "Year"
  ) +
  theme_minimal()
```
