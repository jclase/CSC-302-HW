```r
library(dplyr)
metabolite_file <- "/content/drive/MyDrive/Colab Notebooks/DATA/metabolite.csv"
metabolites_data <- read.csv(metabolite_file, stringsAsFactors = FALSE)
alzheimers_patients <- sum(metabolites_data$Disease == "Alzheimers", na.rm = TRUE)
cat("Number of Alzheimer's patients:", alzheimers_patients, "\n")
missing_values <- colSums(is.na(metabolites_data))
print(missing_values)
cleaned_data <- metabolites_data[!is.na(metabolites_data$Dopamine), ]
median_value <- median(cleaned_data$c4.OH.Pro, na.rm = TRUE)
cleaned_data$c4.OH.Pro[is.na(cleaned_data$c4.OH.Pro)] <- median_value
threshold <- 0.25 * nrow(metabolites_data)
columns_to_keep <- which(colSums(is.na(metabolites_data)) <= threshold)
filtered_data <- metabolites_data[, columns_to_keep]
print(head(filtered_data))
```
