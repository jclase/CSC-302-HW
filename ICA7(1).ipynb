```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "/content/drive/MyDrive/Colab Notebooks/DATA/boxoffice.csv"
boxoffice = pd.read_csv(file_path)
print(boxoffice.head())
star_wars_amount = boxoffice.loc[boxoffice['title_short'] == 'Star Wars', 'amount']
print(f"Box Office Amount for Star Wars: {star_wars_amount.values[0]}")
boxoffice['amount'].plot.bar()
plt.show()
plt.figure(figsize=(15, 3))
plt.bar(boxoffice['title'], boxoffice['amount'])
plt.xticks(rotation=90)
plt.show()
```
