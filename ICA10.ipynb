```python
file_path = "/content/drive/MyDrive/Colab Notebooks/DATA/titanic.csv"
df = pd.read_csv(file_path)
df['Survived'] = df['Survived'].replace({1: 'Survived', 0: 'Died'})
df = df[df['Pclass'] != "?"].copy()
df.loc[:, 'Pclass'] = df['Pclass'].astype(str)
g = sns.FacetGrid(df, row="Pclass", col="Survived", margin_titles=True, height=3, aspect=1.5)
g.map_dataframe(sns.kdeplot, x="Age", hue="Sex", fill=True, alpha=0.4)
titles = [
    "Died in Class 1", "Survived in Class 1",
    "Died in Class 2", "Survived in Class 2",
    "Died in Class 3", "Survived in Class 3"
]
for ax, title in zip(g.axes.flatten(), titles):
    ax.set_title(title)
g.set_axis_labels("Age", "Density")
g.add_legend()
plt.show()
```
