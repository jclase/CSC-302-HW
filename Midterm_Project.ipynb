```python
import pandas as pd
import matplotlib.pyplot as plt

# Correct file path for the Inflation dataset
inflation_path = "/content/drive/MyDrive/Colab Notebooks/Inflation rate 1960 - 2023.csv"

# Load the Inflation Rate dataset
inflation_df = pd.read_csv(inflation_path)

# Convert date column to datetime
inflation_df['observation_date'] = pd.to_datetime(inflation_df['observation_date'])

# Plot Inflation Rate Over Time
plt.figure(figsize=(10, 5))
plt.plot(inflation_df['observation_date'], inflation_df['FPCPITOTLZGUSA'], color='red', label="Inflation Rate (%)")
plt.xlabel("Year")
plt.ylabel("Inflation Rate (%)")
plt.title("Inflation Rate Over Time (1960-2023)")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
inflation_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Inflation rate 1960 - 2023.csv")
unemployment_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/unemployment rate 1948 - 2025.csv")
income_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/median household income 1984-2023.csv")
food_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CPI Food 1947-2025.csv")

# Convert 'observation_date' to datetime in all datasets
for df in [inflation_df, unemployment_df, income_df, food_df]:
    df['observation_date'] = pd.to_datetime(df['observation_date'])

# Rename columns for clarity and consistency
inflation_df.rename(columns={'FPCPITOTLZGUSA': 'Inflation'}, inplace=True)
unemployment_df.rename(columns={'UNRATE': 'Unemployment'}, inplace=True)
income_df.rename(columns={'MEHOINUSA646N': 'Income'}, inplace=True)
food_df.rename(columns={'CPIUFDSL': 'Food_Price'}, inplace=True)

# Merge all four datasets on 'observation_date'
merged_df = pd.merge(inflation_df[['observation_date', 'Inflation']],
                     unemployment_df[['observation_date', 'Unemployment']],
                     on='observation_date', how='inner')

merged_df = pd.merge(merged_df, income_df[['observation_date', 'Income']], on='observation_date', how='inner')
merged_df = pd.merge(merged_df, food_df[['observation_date', 'Food_Price']], on='observation_date', how='inner')

# Drop any rows with missing values (important for plotting)
merged_df.dropna(inplace=True)

# Normalize data for comparison (optional but helpful)
normalized_df = merged_df.copy()
normalized_df.set_index('observation_date', inplace=True)
normalized_df = (normalized_df - normalized_df.min()) / (normalized_df.max() - normalized_df.min())

# Plot all four variables (normalized) on one chart
plt.figure(figsize=(14, 6))
for column in normalized_df.columns:
    plt.plot(normalized_df.index, normalized_df[column], label=column)

plt.title("Comparison of Inflation, Unemployment, Income, and Food Prices (Normalized)")
plt.xlabel("Year")
plt.ylabel("Normalized Value (0 to 1)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
