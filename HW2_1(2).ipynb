```r
library(dplyr)
world_cup_file <- "/content/drive/MyDrive/Colab Notebooks/DATA/WorldCupMatches.csv"
world_cup_data <- read.csv(world_cup_file, stringsAsFactors = FALSE)
num_rows <- nrow(world_cup_data)
num_cols <- ncol(world_cup_data)
cat("Number of rows:", num_rows, "\n")
cat("Number of columns:", num_cols, "\n")
summary(world_cup_data)
unique_locations <- length(unique(world_cup_data$City))
cat("Number of unique locations:", unique_locations, "\n")
avg_attendance <- mean(world_cup_data$Attendance, na.rm = TRUE)
cat("Average attendance:", avg_attendance, "\n")
home_goals <- world_cup_data %>%
  group_by(Home.Team.Name) %>%
  summarise(Total_Goals = sum(Home.Team.Goals, na.rm = TRUE))
print(home_goals)
avg_attendance_per_year <- world_cup_data %>%
  group_by(Year) %>%
  summarise(Average_Attendance = mean(Attendance, na.rm = TRUE))
print(avg_attendance_per_year)
plot(avg_attendance_per_year$Year, avg_attendance_per_year$Average_Attendance, type="o",
     col="blue", xlab="Year", ylab="Average Attendance", main="Attendance Trend Over the Years")
```
